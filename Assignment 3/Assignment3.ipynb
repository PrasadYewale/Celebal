```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

diamonds = sns.load_dataset("diamonds")
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='carat', bins=30, kde=True, color='teal')
plt.title('Distribution of Carat Size')
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=diamonds, x='cut', y='price', palette='pastel')
plt.title('Price Distribution Across Cut Quality')
plt.xlabel('Cut Quality')
plt.ylabel('Price (USD)')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=diamonds, x='clarity', y='price', palette='muted')
plt.title('Price vs Clarity (Violin Plot)')
plt.xlabel('Clarity')
plt.ylabel('Price')
plt.xticks(rotation=30)
plt.show()
```

```python
# Count Plot
plt.figure(figsize=(10, 5))
sns.countplot(data=diamonds, x='color', palette='coolwarm')
plt.title('Count of Diamonds by Color')
plt.xlabel('Color Grade')
plt.ylabel('Count')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(10, 5))
avg_price_by_cut = diamonds.groupby('cut')['price'].mean().reset_index()
sns.barplot(data=avg_price_by_cut, x='cut', y='price', palette='Blues_d')
plt.title('Average Price by Cut Quality')
plt.xlabel('Cut')
plt.ylabel('Average Price')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamonds.sample(1000), x='carat', y='price', hue='cut', palette='Set2', alpha=0.7)
plt.title('Carat vs Price (Sample of 1000 Points)')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.legend(title='Cut')
plt.show()
```

```python
# KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=diamonds, x='price', shade=True, color='darkorange')
plt.title('KDE Plot of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()
```

```python
# Heatmap
correlation_data = diamonds[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_data, annot=True, cmap='viridis', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Pair Plot
sample_diamonds = diamonds.sample(500, random_state=42)
sns.pairplot(sample_diamonds, hue='cut', vars=['carat', 'depth', 'price'], palette='husl')
plt.suptitle('Pairwise Relationship of Carat, Depth & Price by Cut', y=1.02)
plt.show()
```

```python
# Swarm Plot
plt.figure(figsize=(12, 6))
sns.swarmplot(data=diamonds.sample(300), x='clarity', y='carat', hue='cut', palette='Spectral')
plt.title('Carat Distribution by Clarity and Cut (Sample of 300)')
plt.xlabel('Clarity')
plt.ylabel('Carat')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=30)
plt.show()
```
